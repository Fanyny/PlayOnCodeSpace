```python
# Ex1

def nb_occurrences(c, mot):
    """
    retourne le nombre d'occurences de c dans mot
    """
    assert type(c) == str and type(mot) == str
    compteur = 0
    for lettre in mot:
        if lettre == c:
            compteur += 1 #compteur = compteur + 1
    return compteur

#tests
assert nb_occurrences("o", "bonjour") == 2
assert nb_occurrences("a", "abracadabra") == 5
assert nb_occurrences("i", "abracadabra") == 0
print("success")
```

```python
def repetition(n, ch):
    """
    fonction qui retourne le concatenation de ch.
    Le nombre repétitons est n.
    """
    assert type(n) == int and type(ch) == str
    reponse = ""
    for nombre in range(n):
        reponse += ch
    return reponse

#tests
assert repetition(3, "bla") == "blablabla"
assert repetition(2, "cou") == "coucou"
assert repetition(0, "Azerty") == ""
print("succes")
```
